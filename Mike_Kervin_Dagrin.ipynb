```
#titanicdataset assignement by MIKE KERVIN DAGRIN

#Eportation of the libraries that we are going to use for the assignement
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
#Reading of our csv file
titanic_data = pd.read_csv ('TitanicData.csv')
titanic_data.head()
```

```
#use of the info function
titanic_data.info()
```

```
#Use of the describe function
titanic_data.describe()
```

```
# Display summary statistics for categorical columns
titanic_data.describe(include=['O'])
```

```
#Data filtering
#first class and over 50
query = (titanic_data['Pclass'] == 1) & (titanic_data['Age'] > 50)
first_class_over_50 = titanic_data[query]
first_class_over_50
```

```
#data filtering
#Female and under 18
query = (titanic_data['Sex'] == 'female') & (titanic_data['Age'] < 18)
female_under_18 = titanic_data[query]
female_under_18
```

```
#data filtering
#embarked as C and fare over 50
query = (titanic_data['Embarked'] == 'C') & (titanic_data['Fare'] > 50)
embarked_c_over_50 = titanic_data[query]
embarked_c_over_50
```

```
query = (titanic_data['Age'] >= 30) & (titanic_data['Age'] <= 40) & (titanic_data['Parch'] == 0)
age_30_40_no_parch = titanic_data[query]
age_30_40_no_parch
```

```
# male passengers paid less than 10 for their fare?
#creation of the query
query = (titanic_data['Sex'] == 'male') & (titanic_data['Fare'] < 10)
male_paid_less_than_10 = titanic_data[query]
male_paid_less_than_10
```

```
#Average age of passengers by passenger class?
average_age_by_class = titanic_data.groupby('Pclass')['Age'].mean()
average_age_by_class
```

```
#total fare for each embarkation station
average_age_by_class = titanic_data.groupby('Embarked')['Fare'].sum()
average_age_by_class
```

```
#number of passenger in each class
passenger_count_by_class = titanic_data.groupby('Pclass')['PassengerId'].count()
passenger_count_by_class
```

```
#average of fare by male and fenmale
median_fare_by_sex = titanic_data.groupby('Sex')['Fare'].median()
median_fare_by_sex
```

```
#Data visualization
#Age distribution
plt.figure(figsize=(9, 6))
plt.hist(titanic_data['Age'], bins=20, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```
plt.figure(figsize=(9, 6))
plt.hist(titanic_data['Fare'], bins=20, edgecolor='black')
plt.title('Fare Distribution')
plt.xlabel('Fare')

plt.ylabel('Frequency')
plt.show()
```

```
#age distribution by sex
male_age = titanic_data[titanic_data['Sex'] == 'male']['Age']
female_age = titanic_data[titanic_data['Sex'] == 'female']['Age']
plt.hist(male_age, bins=20, alpha=0.4, label='Male')
plt.hist(female_age, bins=20, alpha=0.5, label='Female')
plt.title('Age Distribution by Sex')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```
# Boxplot of Fare by Passenger Class

sns.boxplot(x='Pclass', y='Fare', data=titanic_data)
plt.title('Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```
# Pairplot of selected features
sns.pairplot(titanic_data, vars=['Age', 'Fare', 'Pclass'], hue='Sex')
plt.show()
```

```
#Violin plot of Age distribution by Sex and Passenger Class
sns.violinplot(x='Pclass', y='Age', hue='Sex', data=titanic_data, split=True)
plt.title('Age Distribution by Sex and Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```
# boxplot to visualize Fare by passenger class and sex, explain which groups paid higher fare/ class.
sns.boxplot(x='Pclass', y='Fare', hue= 'Sex', data=titanic_data)
plt.title('Fare by Passenger Class and Sex')
plt.xlabel('Passenger Class and Sex')
plt.ylabel('Fare')
plt.show()
#group that paid higher by class
#Class 1 = Female
#Class 2 = Female
#class 3 = Female
```

```
#Male and female fare who paid over 100
query = (titanic_data['Fare'] > 100)
male_paid_less_than_10 = titanic_data[query]
male_paid_less_than_10
```

```
#representation of passenger who paid more than 100
sns.boxplot(x='Sex', y='Fare', data = titanic_data[titanic_data['Fare'] > 100])
plt.title('Graphical repersentation')
plt.xlabel('')
plt.ylabel('Fare')
plt.show()
```
