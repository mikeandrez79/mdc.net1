```python
# Housing prices prediction assignement.
# 1-Importing the libraries that we need to do the assignement.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, root_mean_squared_error
from sklearn.tree import export_graphviz
from sklearn.tree import export_graphviz
import graphviz
```

```python
# 2-data exploration to understand the available features
datahousing = pd.read_csv('housing.csv')
datahousing = pd.read_csv ('housing.csv')
```

```python
# Exploring the csv file
datahousing.info()
```

```python
datahousing.head(20)
```

```python
# Convert specified columns to dummy variables or booleen values.
columns_to_dummy = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus']
dframe = pd.get_dummies(datahousing, columns=columns_to_dummy, drop_first= False)
dframe.head(10)
```

```python
# Diffining X and Y.
X = dframe.drop(['price'], axis=1)
y = dframe['price']
```

```python
# Splitting our data into two categories test and train.
# Our train data will receive will receive 65% and the data 35%.
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.35, random_state = 42)
```

```python
# Display the shapes of the training and testing datasets
print(f"the shape of the training set (input) is: {X_train.shape}")
print(f"the shape of the training set (target) is: {Y_train.shape}\n")
print(f"the shape of the test set (input) is: {X_test.shape}")
print(f"the shape of the test set (target) is: {Y_test.shape}")
```

```python
#creating a regression decision tree model
# Create a regression tree model
reg_tree = DecisionTreeRegressor(max_depth=3)
```

```python
#training the model
reg_tree.fit(X_train, Y_train)
```

```python
#Prediction
y_pred = reg_tree.predict(X_test)
mse = mean_squared_error(Y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
# Visualize the tree (optional, requires graphviz)
from sklearn.tree import export_graphviz
import graphviz
```

```python
feature_names = X.columns
dot_data = export_graphviz(
    reg_tree, out_file=None, feature_names=feature_names, filled=True, rounded=True, special_characters=True
)
graph = graphviz.Source(dot_data)
graph.render("regression_tree")
```
